```python
import torch
import paddle
paddle.set_device("cpu")
```

```python
import numpy as np
def compare(a, b):
    a = a.numpy()
    b = b.numpy()
    print("mean dif:", np.abs(a - b).mean())
    print("max dif:", np.abs(a - b).max())
```

```python
logits1 = torch.load("logits.pt",map_location="cpu")
logits2 = paddle.load("logits.pd")
compare(logits1,logits2)
```

```python
hidden_states1 = torch.load("hidden_states.pt",map_location="cpu")
hidden_states2 = paddle.load("hidden_states.pd")
for a, b in zip(hidden_states1, hidden_states2):
    compare(a, b)
    print("~" * 50)
```
